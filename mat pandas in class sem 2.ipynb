```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def practicing_line_plot():
    print('hello')
    df=pd.DataFrame({'xvalues': range(0,101), 'yvalues': range(0,101)})
    print(df)
    #plt.plot( 'xvalues', 'yvalues', data=df)
    plt.plot( df.xvalues, df.yvalues)
    plt.show()
```

```python
df2=pd.DataFrame({'xvalues': range(0,100),'yvalues': np.random.randn(100)})
print(df2)
plt.plot('xvalues', 'yvalues', data=df2)
plt.show()
```

```python
years =list(range(1950,2011,10))
gdp =[300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]
plt.plot(years,gdp,color='red',marker='o',linestyle='solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.xlabel("Year")
plt.show()
```
